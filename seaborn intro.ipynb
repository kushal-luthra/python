```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

np.random.seed(sum(map(ord,"aesthetics")))

def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip,label='i ='+str(i))
    plt.grid(False)
    plt.legend(loc=4)

#this will print with default matplotlib
sinplot()
```

```python
#to print with seaborn default import seaborn module
#this will automatically override the matplotlib default settings
import seaborn as sns 
sinplot() 
```

```python
#styling done using: axes_style() and set_style()
#grid themes : 5 default options- whitegrid, darkgrid, dark, white and ticks
#default grid theme is darkgrid
sns.set_style('whitegrid') 
sinplot()
```

```python
#despine() to remove spines from top and right
sinplot()
sns.despine() 
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(121)
    sinplot()
plt.subplot(122)
sinplot(-1)
```

```python
#to view default settings of seaborn
sns.axes_style()
```

```python
#to change style-
sns.set_style("darkgrid" ,{"axes.facecolor":'0.8'}) #value passed as key value pair
sinplot()
```

```python
#to reset the default parameters
sns.set()
```

```python
#The four preset contexts, in order of relative size, are paper, notebook, talk, and poster. 
#The notebook style is the default
sns.set_context("paper")
plt.figure(figsize=(8,6)) #bigger figure
sinplot()
```
